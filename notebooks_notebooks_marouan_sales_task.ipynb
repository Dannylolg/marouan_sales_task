```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

month = np.array(['Jan', 'Feb', 'Mar'])
product_A = np.array([120, 130, 125])
product_B = np.array([150, 160, 170])
product_C = np.array([100, 110, 105])

df = pd.DataFrame({
    'Month' : month,
    'Product_A': product_A,
    'Product_B': product_B,
    'Product_C': product_C
    
})

print(df)
df.describe()
```

### mean means the average.
### min means the minimum number.
### count means how many items or numbers do we have in the array.

```python
plt.plot(
    month, product_A,
    color="red",
    label='Product_A',
    marker="o",
    markerfacecolor="white",
    markeredgecolor="black",
    linestyle="--"
)

plt.plot(
    month, product_B,
    label='Product_B',
    marker="o",
    markerfacecolor="white",
    markeredgecolor="black",
    color="green",
    linestyle="-"
)

plt.plot(
    month, product_C,
    label='Product_C',
    marker="o",
    markerfacecolor="white",
    markeredgecolor="black",
    color="blue",
    linestyle="dashdot"
)

plt.legend()
plt.title("Monthly sales data")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True, alpha=0.7)
plt.show()
```

### The product_A chart (red) shows more sales in February but less sales in March.
### The product_B chart (green) shows more sales in February and more sales in March.
### The product_C chart (blue) shows more sales in February and less sales in March.

```python
m_valueA = np.median(product_A)
m_valueB = np.median(product_B)
m_valueC = np.median(product_C)
print(m_valueA)
print(m_valueB)
print(m_valueC)
```

```python
labels = ['Product A', 'Product B', 'Product C']
medians = [m_valueA, m_valueB, m_valueC]

plt.figure(figsize=(8, 5))

# Plotting each point individually to match your previous styling
plt.scatter(labels[0], medians[0], color="red", marker="o", s=100, edgecolors="black", label="Median A")
plt.scatter(labels[1], medians[1], color="green", marker="s", s=100, edgecolors="black", label="Median B")
plt.scatter(labels[2], medians[2], color="blue", marker="D", s=100, edgecolors="black", label="Median C")

plt.title("Median Sales per Product")
plt.ylabel("Median Sales Value")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
